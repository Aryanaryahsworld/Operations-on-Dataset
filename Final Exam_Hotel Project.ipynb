And we noticed that the hotelName is conbined by the number and hotel name. We need to process it before further action.  And the `hotelUrl` is the unique url for each hotel. We can ignore it.

```python
def parse_hotel_number_and_name(text):
    lines = text.strip().split('\n')
    num, name = lines[0].split(None, 1)
    return [num, name]

hotel_reviews['hotel_num'] = hotel_reviews.hotelName.apply(lambda x: parse_hotel_number_and_name(x)[0])
hotel_reviews['hotel_name'] = hotel_reviews.hotelName.apply(lambda x: parse_hotel_number_and_name(x)[1])
hotel_reviews.drop(['hotelName', 'hotelUrl'], axis=1, inplace=True)
hotel_reviews
```

# Clustering
Our idea of clustering hotels is based on its reviews.
1. Build a topic model for the reviews, and generate topic results for each of them.
2. Count how many reviews of each topic does each hotel have.
3. Cluster the hotels based on their reviews' topics counts.

## Topic Modeling - Latent Dirichlet Allocation

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
 
nltk.download('stopwords')
 
from nltk.corpus import stopwords
```

```python
reviews = hotel_reviews['review_body']
reviews
```

Before do the topic modeling, we need to do some preprocess to the reviews:

1. Tokenization: Split the text into sentences and the sentences into words.    
1. Lowercase the words and remove punctuation.
1. All stopwords are removed.
1. Words are lemmatized — words in third person are changed to first person and verbs in past and future tenses are changed into present.
1. Words are stemmed — words are reduced to their root form.

```python
stopwords = stopwords.words('english')


def tokenization_and_stemming(text):
    '''
    INPUT
    text - string
    OUTPUT
    clean_tokens - a list of words
    This function processes the input using the following steps :
    1. Remove punctuation characters
    2. Tokenize text into list
    3. Stem, Normalize and Strip each word
    4. Remove stop words
    '''
    # Remove punctuation characters and numbers
    text = re.sub(r"[^a-zA-Z]", " ", text)
    
    # Tokenize text
    tokens = word_tokenize(text)
    

    # Create a instance of stem class
    stemmer = SnowballStemmer("english")
    
    clean_tokens = []
    for word in tokens:
        clean_tok = stemmer.stem(word).lower().strip()
        if clean_tok not in stopwords:
            clean_tokens.append(clean_tok)

    return clean_tokens
```

Then we use CountVectorizer to create bag of words. 

```python
# Term frequency for LDA model
tf_lda = CountVectorizer(
    max_df=0.99,
    max_features=500,
    min_df=0.01,
    tokenizer=tokenization_and_stemming,
    ngram_range=(1,1))


tf_matrix_lda = tf_lda.fit_transform(reviews) 


print ("In total, there are {} reviews and {} terms.".format(
    str(tf_matrix_lda.shape[0]), str(tf_matrix_lda.shape[1])
))
```

Now we can use LDA to create topics from the tf_matrix. We will split these reviews into 3 topics.

```python
# Use LDA for clustering
LDA = LatentDirichletAllocation(n_components=3)


# Document topic matrix for tf_matrix_lda
lda_output = LDA.fit_transform(tf_matrix_lda)
print(lda_output.shape)
```

```python
# Topics and words matrix
# Components_[i, j] can be viewed as pseudocount that represents the number of times word j was assigned to topic i.
topic_word = LDA.components_
print(topic_word.shape)
```

Now we get the values of these 3 topics for each review. And we choose the topic with highest score as The Topic for the review.

```python
# Column names
topic_names = ["Topic" + str(i) for i in range(LDA.n_components)]

# Index names
doc_names = [i for i in range(len(reviews))]

df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topic_names, index=doc_names)

# Get dominant topic for each document
topic = np.argmax(df_document_topic.values, axis=1)
df_document_topic['Topic'] = topic

df_document_topic
```

```python
df_document_topic['Topic'].value_counts()
```

From above we can see, topic 0 has 615 reviews and topic 1 has 592 reviews. topic 2 has 552 reviews. They are roughly even splited.   
Next, we need to merge the topic results to original review data.

```python
hotel_reviews['Topic'] = df_document_topic['Topic']
hotel_reviews
```

## Topic counts
Now we can count how many topics does each hotel have.

```python
topic_counts = pd.pivot_table(
    data=hotel_reviews[['hotel_num','Topic']], 
    index='hotel_num', 
    columns='Topic', 
    aggfunc=len, 
    fill_value=0
)
topic_counts.columns = [f'topic{i}' for i in range(topic_counts.shape[1])]
topic_counts.head(5)
```

From above we can see, hotel 0 has 13 reviews on topic 0 , 26 reviews on topic 1 and 15 reviews on topic 2.

## Clustering Hotels

```python
from sklearn.cluster import KMeans
km = KMeans()
```

```python
# Elbow Method 
WCSS = []
for i in range(1,11):
  km= KMeans(n_clusters=i)
  km.fit(topic_counts)
  WCSS.append(km.inertia_)

plt.plot(range(1,11),WCSS,marker='o',linestyle='--',color='r')
```

From the above elbow chart, we choose 4 as our cluster number.

```python
  km= KMeans(n_clusters=4)
  km.fit(topic_counts)

  topic_counts['group'] = km.labels_
  topic_counts.head(5)
```

We can see hotel 0, 1, 10, 11 are in the same group 1, while hotel 12 is in group 3.

After installation, we need to create a SentenceTransformer model

```python
from sentence_transformers import SentenceTransformer
# initialize sentence transformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

## Create Vectors from Reviews  
Now, use the sentencetransfromer to encode the review text into vectors.

```python
reviews = hotel_reviews['review_body'].tolist()
```

```python
%%time
review_embeddings = model.encode(reviews)
review_embeddings.shape
```

## Build FAISS Index   
After we get the vectors, we can use FAISS to create an L2 distance index. 

```python
import faiss

vector_dimension = review_embeddings.shape[1]
index = faiss.IndexFlatL2(vector_dimension)
faiss.normalize_L2(review_embeddings)
index.add(review_embeddings)
```

## Search and Results
To seach the hotel, first we need to encode the seach text, and use FAISS to search the nearest reviews. From these reviews, we can get the recommended hotel.

```python
# encode the seach text
search_text = 'Cheap hotel near downtown'
search_vector = model.encode(search_text)
_vector = np.array([search_vector])
faiss.normalize_L2(_vector)
```

```python
# Search for the 10 nearest reviews indexes 
k = 10
D, I = index.search(_vector, k=k)
```

```python
print(D, I)
```

After get the indexes, we can print the recommended hotels.

```python
print(f'The seach text is "{search_text}"')
print(f'The recommended hotels are : ')
for i in I[0]:
  print(f'Hotel : {hotel_reviews.hotel_name[i]}')
  print(f'Related Reviews: {reviews[i]}')
  print('-'*220)
```

# **Discussion**

In this project,we tried out both K-means and LDA as examples of clustering and topic modeling.

**K-means** has some limitations. It is very sensitive to outliers. As you may have already seen, it can produce very small clusters corresponding to outliers. And K-means also has difficulties with clusters of different sizes and densities. That is why we randomly select equal size subsets of different eviews from the whole dataset.

**Latent Dirichlet allocation (LDA)** is a generative statistical model that allows sets of observations to be explained by unobserved groups that explain why some parts of the data are similar. The LDA model is highly modular and can, therefore, be easily extended. The main field of interest is modeling relations between topics. In this task, LDA did a better job of clustering the reviews.    

Also, we did a review based hotel search using FAISS.   
To search in the reviews, we created vectors for each of them and calculated their cosine distances.     

**FAISS** is a library that allows developers to quickly search for embeddings of multimedia documents that are similar to each other. It solves limitations of traditional query search engines that are optimized for hash-based searches, and provides more scalable similarity search functions.
