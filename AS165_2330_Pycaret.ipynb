```python
#apply pycaret to the customer churn data and do a small write up on which model performed best
# install pycaret

# import libraries
import pandas as pd
import numpy as np
# read csv data
data = pd.read_csv('https://raw.githubusercontent.com/coen2812023/Classification/main/customer_churn.csv')
# check data types
data.dtypes
# replace blanks with np.nan
data['balance'] = data['balance'].replace(' ', np.nan)
# convert to float64
data['balance'] = data['balance'].astype('float64')
```

```python
import plotly.express as px

fig = px.scatter(x=data['tenure'], y=data['TotalCharges'], 
                 color = data['Churn'], template = 'presentation', 
                 opacity = 0.5, facet_col = data['Contract'], 
                 title = 'Customer Churn by Tenure, Charges, and Contract Type',
                 labels = {'x' : 'Customer Tenure', 'y' : 'Total Charges $'})
                
fig.show()
```
